### basic string functionalities

```python
print(my_name)
my_name_lower=my_name.lower() #converting to lower case
print(my_name_lower)
my_name_upper=my_name.upper()  #converting to upper case
print(my_name_upper)
print(my_name.swapcase()) # swaping the lower/upper
print(my_name_lower.capitalize()) # converting 1st letter in upper case
```

## arthametic operators


```python
num_1=2026
num_2=7

num_addition=num_1+num_2  # 2026+7
num_subtraction=num_1-num_2  #2026-7
num_multiplication=num_1*num_2
num_division=num_1/num_2
print('result of addition','\t', '=',num_addition)
print('result of subtraction','\t', '=',num_subtraction)

num_modulus=num_1 % num_2  #reminder
num_exponent=num_1**num_2  # 2026^7
num_floor_division=num_1//num_2
print('result of modulus', '\t', '=',num_modulus)
print('result exponent', '\t', '=',num_exponent)
print('result of floor division', '\t', '=',num_floor_division)
```
